## (a) Math

### i. Prove vectorized equation

Assume row of data is `m` and feature of data is `n`, and $w^{(i)}$ is shaped $1 \times m$

$$\begin{align*}
\text{Let } \underbrace{B^{(i)}}_{scaler} &= \theta^T x^{(i)} - y^{(i)} \text{, then} \\\\
J(\theta) &= \frac{1}{2} \sum^m_{i=1} w^{(i)} \left( \theta^T x^{(i)} - y^{(i)} \right)^2 \\\\
&= \frac{1}{2} \sum^m_{i=1} w^{(i)} (B^{(i)})^2 \\\\
\text{For } J(\theta) &= \underbrace{(X\theta -y)^T}_{1 \times m} \underbrace{W}_{m \times m} \underbrace{(X\theta -y)}_{m \times 1}, \text{ Let } A = (X\theta -y) \text{, then:} \\\\
J(\theta) &= A^T WA \\\\
&= \sum^m_{i=1} (B^{(i)} w^{(i)}) A \\\\
&= \sum^m_{i=1} ((B^{(i)})^2 w^{(i)}) = \sum^m_{i=1} w^{(i)} (B^{(i)})^2  \\\\
\end{align*}$$

### ii. derive a loss function

$$\begin{align*}
J(\theta) &= (X\theta -y)^T W (X\theta -y) \\\\

&\text{Expanding the expression, we get:} \\\\
&= \theta^TX^TWX\theta - \underbrace{\theta^TX^TWy - y^TWX\theta}_{\text{scaler}} + y^TWy \\\\

&= \theta^TX^TWX\theta - 2\theta^TX^TWy  + y^TWy \\\\

&\text{Now, taking the derivative with respect to $\theta$, we get:}\\\\
\frac{\partial J(\theta)}{\partial \theta} &= 2X^TWX\theta - 2X^TWy + 0\\\\

&\text{Simplifying further, we get:}\\\\
&= 2X^TWX\theta - 2X^TWy \\\\

&\text{Finally, we can factor out $2X^TW$, giving us:} \\\\
&= 2X^TW(X\theta - y) \\\\\\

&\text{To get the optimal parameters, let } \\\\
0 &= 2X^TWX\theta - 2X^TWy \\\\
&\text{then:} \\\\
X^TWX\theta &= X^TWy \\\\
\theta &= (X^TWX)^{-1}X^TWy
\end{align*}$$

#### Note:

__*__ $\frac{\partial (\theta^TA\theta)}{\partial \theta} = 2A\theta$. Reference [from here.](https://en.wikipedia.org/wiki/Matrix_calculus#:~:text=Scalar%2Dby%2Dvector%20identities) Or proof yourself.


### iii. 

$$\begin{align*}
&\text{The likelyhood function can be write as:} \\\\
\mathcal{L} &= \prod^m_{i=1} \frac{1}{\sqrt{2\pi} \sigma^{(i)}} \exp \left(  - \frac{(y^{(i)} - \theta^T x^{(i)})^2}{2(\sigma^{(i)})^2}  \right) \\\\

&\text{After convert to log-likelyhood:} \\\\
\ell &= \log \left[ \prod^m_{i=1} \frac{1}{\sqrt{2\pi} \sigma^{(i)}} \exp \left(  - \frac{(y^{(i)} - \theta^T x^{(i)})^2}{2(\sigma^{(i)})^2}  \right) \right] \\\\
&= \sum^m_{i=1} \log\left[ \frac{1}{\sqrt{2\pi} \sigma^{(i)}} \exp \left(  - \frac{(y^{(i)} - \theta^T x^{(i)})^2}{2(\sigma^{(i)})^2}  \right) \right] \\\\ 
&= \ell = \sum^m_{i=1} \log \left[ \frac{1}{\sqrt{2\pi} \sigma^{(i)}} \exp \left(  - \frac{(y^{(i)} - \theta^T x^{(i)})^2}{2(\sigma^{(i)})^2}  \right) \right] \\\\

&\text{Since: } \log(a/b) = \log(a) - \log(b),  \text{This gives us:} \\\\
&= \sum^m_{i=1} \left[ \log(1) - \log(\sqrt{2\pi} \sigma^{(i)}) + \log\left(\exp \left(  - \frac{(y^{(i)} - \theta^T x^{(i)})^2}{2(\sigma^{(i)})^2}  \right)\right) \right]\\\\

&\text{Since: } \log(1) = 0, \text{we can simplify the expression further:} \\\\
&= \sum^m_{i=1} \left[ -\log(\sqrt{2\pi} \sigma^{(i)}) + \log\left(\exp \left(  - \frac{(y^{(i)} - \theta^T x^{(i)})^2}{2(\sigma^{(i)})^2}  \right)\right) \right] \\\\

&\text{because } \log(exp(x)) = x  \\\\
&= \sum^m_{i=1} \left[ -\log(\sqrt{2\pi} \sigma^{(i)}) +  -\frac{(y^{(i)} - \theta^T x^{(i)})^2}{2(\sigma^{(i)})^2}   \right] \\\\
&= -\sum^m_{i=1}  \left[\log(\sqrt{2\pi} \sigma^{(i)}) +  \frac{(y^{(i)} - \theta^T x^{(i)})^2}{2(\sigma^{(i)})^2}\right] \\\\

&\text{Next, lets find the } \theta \text{ that will minimize the log-likelyhood:} \\\\

\frac{\partial \ell}{\partial \theta} &=  -\sum^m_{i=1}  \left[\log(\sqrt{2\pi} \sigma^{(i)}) +  \frac{(y^{(i)} - \theta^T x^{(i)})^2}{2(\sigma^{(i)})^2}\right] = 0 \\\\

&\Rightarrow \sum^m_{i=1} \log(\sqrt{2\pi} \sigma^{(i)}) + \sum^m_{i=1} \frac{(y^{(i)} - \theta^T x^{(i)})^2}{2(\sigma^{(i)})^2} = 0 \\\\

&\Rightarrow \frac{\partial}{\partial \theta}\left(\sum^m_{i=1} \frac{(y^{(i)} - \theta^T x^{(i)})^2}{2(\sigma^{(i)})^2}\right) = 0 \\\\
&\Rightarrow \sum^m_{i=1} \frac{-x^{(i)} (y^{(i)} - \theta^T x^{(i)})}{(\sigma^{(i)})^2} = 0 \\\\
&\Rightarrow \sum^m_{i=1} \frac{-x^{(i)} y^{(i)}}{(\sigma^{(i)})^2} + \sum^m_{i=1} \frac{x^{(i)} \theta^T x^{(i)}}{(\sigma^{(i)})^2} = 0 \\\\
&\Rightarrow \sum^m_{i=1} \frac{x^{(i)} \theta^T x^{(i)}}{(\sigma^{(i)})^2} = \sum^m_{i=1} \frac{x^{(i)} y^{(i)}}{(\sigma^{(i)})^2} \\\\
&\text{Next, we can factor out } \theta \text{ from the left side of the equation:} \\\\
&\Rightarrow \theta^T \sum^m_{i=1} \frac{(x^{(i)})^2}{(\sigma^{(i)})^2} = \sum^m_{i=1} \frac{x^{(i)} y^{(i)}}{(\sigma^{(i)})^2} \\\\
&\Rightarrow \theta = (\sum^m_{i=1} \frac{(x^{(i)})^2}{(\sigma^{(i)})^2})^{-1} \sum^m_{i=1} \frac{x^{(i)} y^{(i)}}{(\sigma^{(i)})^2} \\\\
\end{align*}$$

We can define a diagonal matrix $W$ where $W_{ii} = \frac{1}{(\sigma^{(i)})^2}$, and a column vector $y$ where $y_i = y^{(i)}$. Similarly, we can define a matrix $X$ where the $i$-th row of $X$ is $x^{(i)T}$.

Substituting these definitions into the equation, we get:
$$\begin{align*}
\theta &= (\sum^m_{i=1} (x^{(i)})^2 w^{(i)})^{-1} \sum^m_{i=1} x^{(i)} y^{(i)} w^{(i)} \\\\
&= (X^TWX)^{-1}X^TWy
\end{align*}$$

## (b) Code

```python
import matplotlib.pyplot as plt
import numpy as np
import src.util as util

from src.linear_model import LinearModel
```

```python
def mean_squared_error(y_true, y_pred):
    mse = np.mean((y_true - y_pred) ** 2)
    return mse

def plot_lr(x, y_true, y_pred):
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(x[:, -1], y_true, 'bx',label= "true")
    plt.plot(x[:, -1], y_pred, 'ro',label="pred")
    plt.legend()
    plt.show()


class LocallyWeightedLinearRegression(LinearModel):
    """Locally Weighted Regression (LWR).

    Example usage:
        > clf = LocallyWeightedLinearRegression(tau)
        > clf.fit(x_train, y_train)
        > clf.predict(x_eval)
    """

    def __init__(self, tau):
        super(LocallyWeightedLinearRegression, self).__init__()
        self.tau = tau
        self.x = None
        self.y = None

    def fit(self, x, y):
        """Fit LWR by saving the training set.

        """
        # *** START CODE HERE ***
        self.x = x; self.y = y
        # *** END CODE HERE ***

    def predict(self, x):
        """Make predictions given inputs x.

        Args:
            x: Inputs of shape (m, n).

        Returns:
            Outputs of shape (m,).
        """
        # *** START CODE HERE ***
        # we have to iterate because we have to get new theta for which local interval!!
        m = x.shape[0]
        y_pred = np.zeros(m)
        for i in range(m):
            dist = np.linalg.norm(self.x - x[i], axis=1, ord=2)
            w = np.diag(np.exp(- dist / (2 * self.tau**2)))
            theta = np.linalg.inv(self.x.T @ w @ self.x) @ self.x.T @ w @ self.y
            y_pred[i] = x[i] @ theta
        return y_pred

        # *** END CODE HERE ***


def problem_b(tau, train_path, eval_path):
    """Problem 5(b): Locally weighted regression (LWR)

    Args:
        tau: Bandwidth parameter for LWR.
        train_path: Path to CSV file containing dataset for training.
        eval_path: Path to CSV file containing dataset for evaluation.
    """
    # Load training set
    x_train, y_train = util.load_dataset(train_path, add_intercept=True)
    x_eval, y_eval = util.load_dataset(eval_path, add_intercept=True)
    # *** START CODE HERE ***
    # Fit a LWR model
    model = LocallyWeightedLinearRegression(tau)
    model.fit(x_train, y_train)
    pred = model.predict(x_eval)
    # Get MSE(Mean square error) value on the validation set
    mse = mean_squared_error(y_eval, pred)
    print("mean_squared_error: ", mse)
    # Plot validation predictions on top of training set
    plot_lr(x_eval, y_eval, pred)
    # *** END CODE HERE ***


problem_b(0.5,"./data/ds5_train.csv","./data/ds5_valid.csv") # under fitting!
```
